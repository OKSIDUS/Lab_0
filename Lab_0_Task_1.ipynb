# Норвегія

## Огляд

Норвегія - країна в Північній Європі, відома своєю природою та історією.

### Площа
Площа Норвегії становить приблизно 385 207 квадратних кілометрів.

### Населення
Населення країни нараховує близько 5,408 мільйонів жителів.

## Державний устрій
Норвегія є конституційною монархією з парламентською системою уряду.

## Видатні персоналії

- *Оле Йоханнессен* - відомий норвезький митець.
- **Грета Тунберг** - активістка із захисту навколишнього середовища.

## Пам'ятки культури та природи

### Культурні пам'ятки

1. Музей вікінгів у Осло.
2. Опера в Бергені.

### Природні краси

- *Фьорди Норвегії* - вражаючі природні утворення.
